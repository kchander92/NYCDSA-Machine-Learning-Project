```python
housing2 = housing.drop(columns=['PID', 'YrsLaterBlt', 'YrsAfterRemod', 'GarageYrsAfterBlt',
                                 'YrsAfterSold', 'GeoRefNo', 'Prop_Addr'])
housing2.head()
```

```python
object_cols = housing2.loc[:, (housing2.dtypes == object)].columns
housing2_num = housing2.drop(columns = object_cols)

for col in object_cols:
    dummies = pd.get_dummies(housing2[col], prefix = col)
    housing2_num = pd.concat([housing2_num, dummies], axis = 1)
    
housing2_num.head(10)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor

X = housing2_num.drop(columns='SalePrice')
y = housing2_num['SalePrice']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)

tree = DecisionTreeRegressor()

depth = range(5, 50)
train_R2 = []
test_R2 = []
for i in depth:
    tree.set_params(max_depth=i, min_samples_leaf=1)
    tree.fit(X_train, y_train)
    train_R2.append(tree.score(X_train, y_train))
    test_R2.append(tree.score(X_test, y_test))

import matplotlib.pyplot as plt
plt.plot(depth, train_R2, c='red', label='training R2')
plt.plot(depth, test_R2, c='blue', label='test R2')
plt.legend()
plt.show()
```

```python
plt.plot(depth, np.array(train_R2) - np.array(test_R2), c='red')
```

```python
from sklearn.model_selection import GridSearchCV, cross_val_score, KFold

kfold = KFold(n_splits=5, random_state=42, shuffle=True)

grid_para_tree = [{
#    'criterion': ['squared_error', 'friedman_mse', 'absolute_error', 'poisson'],
#    "max_depth": range(5, 20),
    "min_samples_leaf": range(1, 10),
    "min_samples_split": np.linspace(start=2, stop=30, num=15, dtype=int)   
}]
grid_search_tree = GridSearchCV(
    tree, grid_para_tree, cv=kfold, n_jobs=-1, verbose=3)

grid_search_tree.fit(X, y)
```

```python
grid_search_tree.best_params_
```

```python
grid_search_tree.best_score_
```

```python
best_tree = grid_search_tree.best_estimator_
tree_feature_importances = pd.DataFrame({'features': best_tree.feature_names_in_,
                                         'importances': best_tree.feature_importances_}).\
                           sort_values(by='importances', ascending=False)
tree_feature_importances
```
