```python
housing2['HouseAge'] = housing2['YearBuilt'].max() - housing2['YearBuilt']
housing2.drop(columns='YearBuilt', inplace=True)

housing2['1stFlrRatio'] = housing2['1stFlrSF'] / housing2['GrLivArea']
housing2['2ndFlrRatio'] = housing2['2ndFlrSF'] / housing2['GrLivArea']
housing2.drop(columns=['1stFlrSF', '2ndFlrSF'], inplace=True)

housing2['BsmtFin1Ratio'] = housing2['BsmtFinSF1'] / housing2['TotalBsmtSF']
housing2['BsmtFin2Ratio'] = housing2['BsmtFinSF2'] / housing2['TotalBsmtSF']
housing2['BsmtUnfRatio'] = housing2['BsmtUnfSF'] / housing2['TotalBsmtSF']
housing2.drop(columns=['BsmtFinSF1', 'BsmtFinSF2', 'BsmtUnfSF'], inplace=True)

housing2['GarageRatio'] = housing2['GarageArea'] / housing2['LotArea']
housing2['WoodDeckRatio'] = housing2['WoodDeckSF'] / housing2['LotArea']
housing2['OpenPorchRatio'] = housing2['OpenPorchSF'] / housing2['LotArea']
housing2['EnclosedPorchRatio'] = housing2['EnclosedPorch'] / housing2['LotArea']
housing2['3SsnPorchRatio'] = housing2['3SsnPorch'] / housing2['LotArea']
housing2['ScreenPorchRatio'] = housing2['ScreenPorch'] / housing2['LotArea']
housing2['PoolAreaRatio'] = housing2['PoolArea'] / housing2['LotArea']
housing2.drop(columns=['GarageArea', 'WoodDeckSF', 'OpenPorchSF',
                       'EnclosedPorch', '3SsnPorch', 'ScreenPorch', 'PoolArea'], inplace=True)

housing2.fillna(0, inplace=True)
```

```python
housing2[housing2['LotArea'] > 50000]
```

```python
from sklearn.preprocessing import OrdinalEncoder

categorical_features = housing2.loc[:, (housing2.dtypes == object)].columns
ordered_features = ['ExterQual', 'ExterCond', 'BsmtQual', 'BsmtCond', 'HeatingQC',
                    'KitchenQual', 'FireplaceQu', 'GarageQual', 'GarageCond']

encoders = {}

for feature in categorical_features:
    if feature in ordered_features:
        oe = OrdinalEncoder(categories = [['None', 'Po', 'Fa', 'TA', 'Gd', 'Ex']])
    else:
        oe = OrdinalEncoder()
    
    housing2[[feature]] = oe.fit_transform(housing2[[feature]])
    encoders[feature] = oe
```

```python
encoders['Alley'].categories_
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import GridSearchCV, cross_val_score, KFold

kfold = KFold(n_splits=5, random_state=42, shuffle=True)
X = housing2.drop(columns='SalePrice')
y = housing2['SalePrice']

tree = DecisionTreeRegressor()
grid_para_tree = [{
#    'criterion': ['squared_error', 'friedman_mse', 'absolute_error', 'poisson'],
    "max_depth": range(5, 20),
    "min_samples_leaf": range(1, 10),
    "min_samples_split": np.linspace(start=2, stop=30, num=15, dtype=int)   
}]
grid_search_tree = GridSearchCV(
    tree, grid_para_tree, cv=kfold, n_jobs=-1, verbose=3)

grid_search_tree.fit(X, y)
```

```python
grid_search_tree.best_params_
```

```python
grid_search_tree.best_score_
```

```python
best_tree = grid_search_tree.best_estimator_
tree_feature_importances = pd.DataFrame({'features': best_tree.feature_names_in_,
                                         'importances': best_tree.feature_importances_}).\
                           sort_values(by='importances', ascending=False)
tree_feature_importances
```

```python
from sklearn.model_selection import train_test_split

train_scores = []
test_scores = []
y = housing2['SalePrice']

gradboost = GradientBoostingRegressor()
gradboost.set_params(learning_rate = grid_search_gradboost.best_params_['learning_rate'],
                     min_samples_leaf = grid_search_gradboost.best_params_['min_samples_leaf'],
                     min_samples_split = grid_search_gradboost.best_params_['min_samples_split'],
                     n_estimators = grid_search_gradboost.best_params_['n_estimators'])

for i in range(1, len(gradboost_best_features) + 1):
    feature_subset = list(gradboost_best_features[:i])
    X = housing2[feature_subset]
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3,
                                                        random_state=42)
    
    gradboost.fit(X_train, y_train)
    train_scores.append(gradboost.score(X_train, y_train))
    test_scores.append(gradboost.score(X_test, y_test))
```
