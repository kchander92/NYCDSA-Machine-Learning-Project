```python
housing_clean['YrsLaterBlt'] = housing_clean['YearBuilt'] - housing_clean['YearBuilt'].min()
housing_clean['YrsAfterRemod'] = housing_clean['YearRemodAdd'] - housing_clean['YearBuilt']
housing_clean['GarageYrsAfterBlt'] = housing_clean['GarageYrBlt'] - housing_clean['YearBuilt']
housing_clean['YrsAfterSold'] = housing_clean['YrSold'] - housing_clean['YearBuilt']
housing_clean.head(10)
```

```python
def coords(address):
    try:
        location = geolocator.geocode(address)
        return (location.latitude, location.longitude)
    except:
        return np.nan
```

```python
housing_clean2['Prop_Addr']
```

```python
import re

housing_clean2['Prop_Addr'] = housing_clean2['Prop_Addr'].apply(lambda s: re.sub(r' \d+$', '', s)).map('{} Ames'.format)
housing_clean2['Prop_Addr']
```

```python
housing_clean2['locations'] = housing_clean2['Prop_Addr'].apply(coords)
```

```python
housing_clean2['locations']
```

```python
housing_clean2['locations'][0][0]
```

```python
housing_clean2.dropna(subset='locations', inplace=True)
housing_clean2.isna().sum()
```

```python
housing_clean2['latitude'] = housing_clean2['locations'].apply(lambda coords: coords[0])
housing_clean2['longitude'] = housing_clean2['locations'].apply(lambda coords: coords[1])
housing_clean2.drop(columns='locations', inplace=True)
housing_clean2.head(10)
```

```python
housing_clean2['latitude'] == housing_clean2['latitude'].min()
```

```python
housing_clean2 = housing_clean2[housing_clean2['latitude'] != housing_clean2['latitude'].min()]
housing_clean2
```

```python
plt.boxplot(housing_clean2['latitude'])
```

```python
from sklearn.svm import SVR

svr = SVR(gamma=10)
svr.fit(housing_clean2[['latitude', 'longitude']], housing_clean2['YrsLaterBlt'])
svr.score(housing_clean2[['latitude', 'longitude']], housing_clean2['YrsLaterBlt'])
```

```python
gammas = np.logspace(-10, 5, 150)
gamma_scores = []
for gamma in gammas:
    svr.set_params(gamma = gamma)
    svr.fit(housing_clean2[['latitude', 'longitude']], housing_clean2['YrsLaterBlt'])
    gamma_scores.append(svr.score(housing_clean2[['latitude', 'longitude']], housing_clean2['YrsLaterBlt']))
gamma_scores = np.array(gamma_scores)  

plt.plot(gammas, gamma_scores)
```

```python
[gammas[np.argmax(gamma_scores)], np.max(gamma_scores)]
```
