```python
from sklearn.linear_model import LinearRegression

lm = LinearRegression()
lm.fit(housing_clean[['YrsLaterBlt']], np.log(housing_clean['SalePrice']))
print((lm.coef_, lm.intercept_))
lm.score(housing_clean[['YrsLaterBlt']], np.log(housing_clean['SalePrice']))
```

$\ln P = -0.00749 Y + 12.334$  
$P = \exp(12.334) \times \exp(-0.00749Y) = 227294 \times 0.992^{Y}$

```python
plt.yscale('log')

sb.scatterplot(data=housing_clean, x='YrsLaterBlt', y='SalePrice', alpha=0.2)
plt.plot(range(0, 140), np.exp(lm.intercept_)*np.power(np.exp(lm.coef_[0]),range(0, 140)), 'r')
```

```python
from sklearn.model_selection import train_test_split
from sklearn.svm import SVR

X_train, X_test, y_train, y_test = train_test_split(housing_clean2[['latitude', 'longitude']],
                                                    housing_clean2['SalePrice'],
                                                    test_size=0.3)

svr = SVR()
gammas = np.logspace(0, 5, 10)
scores = []
for gamma in gammas:
    svr.set_params(gamma=gamma)
    svr.fit(X_train,y_train)
    scores.append(svr.score(X_train,y_train)-svr.score(X_test,y_test)) # train R^2 and test R^2 gaps
scores = np.array(scores)   
```
