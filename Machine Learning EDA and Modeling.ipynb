```python
import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
%matplotlib inline

housing = pd.read_csv('Ames_HousePrice.csv', index_col=0)
housing.shape
```

```python
pd.set_option('display.max_columns', None)
```

```python
housing.head()
```

```python
housing.columns
```

```python
housing.index
```

```python
na_vals = housing.isna().mean().sort_values(ascending=False) * 100
na_vals[na_vals > 0]
```

```python
housing_clean = housing.copy()
housing_clean.drop(columns = list(na_vals[na_vals > 5].index), inplace=True)
housing_clean.head()
```

```python
for col in housing_clean.loc[:, (housing_clean.dtypes == object)].columns:
    print(housing_clean.loc[:, (housing_clean.dtypes == object)][col].value_counts())
```

```python
housing_clean.hist(figsize=(16, 16));
```

```python
housing_clean[[col for col in housing_clean.columns if 'Year' in col or 'Yr' in col]].max()
```

```python
housing_clean['YrsAgoBlt'] = housing_clean['YearBuilt'].max() - housing_clean['YearBuilt']
housing_clean['YrsAfterRemod'] = housing_clean['YearRemodAdd'] - housing_clean['YearBuilt']
housing_clean['GarageYrsAfterBlt'] = housing_clean['GarageYrBlt'] - housing_clean['YearBuilt']
housing_clean['YrsAfterSold'] = housing_clean['YrSold'] - housing_clean['YearBuilt']
housing_clean.head(10)
```

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(14, 10), sharey=True)
#plt.xscale('log')
#plt.yscale('log')

sb.scatterplot(data=housing_clean, x='YrsAgoBlt', y='SalePrice', ax=ax1)
sb.scatterplot(data=housing_clean, x='YrsAfterRemod', y='SalePrice', ax=ax2)
sb.scatterplot(data=housing_clean, x='GarageYrsAfterBlt', y='SalePrice', ax=ax3)
sb.scatterplot(data=housing_clean, x='YrsAfterSold', y='SalePrice', ax=ax4)
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(14, 4), sharey=True)

sb.scatterplot(data=housing_clean, x='YrsAfterRemod', y='YrsAgoBlt', ax=ax1)
sb.scatterplot(data=housing_clean, x='GarageYrsAfterBlt', y='YrsAgoBlt', ax=ax2)
sb.scatterplot(data=housing_clean, x='YrsAfterSold', y='YrsAgoBlt', ax=ax3)
```

```python
sb.scatterplot(data=housing_clean, x='YearBuilt', y='YrSold')
```

```python
plt.yscale('log')

sb.scatterplot(data=housing_clean, x='YrsAgoBlt', y='SalePrice')
```

```python
from sklearn.linear_model import LinearRegression

lm = LinearRegression()
lm.fit(housing_clean[['YrsAgoBlt']], np.log(housing_clean['SalePrice']))
print((lm.coef_, lm.intercept_))
lm.score(housing_clean[['YrsAgoBlt']], np.log(housing_clean['SalePrice']))
```

$\ln P = -0.00749 Y + 12.334$  
$P = \exp(12.334) \times \exp(-0.00749Y) = 227294 \times 0.992^{Y}$

```python
sb.scatterplot(data=housing_clean, x='YrsAgoBlt', y='SalePrice', alpha=0.2)
plt.plot(range(0, 130), 227294*np.power(0.992,range(0, 130)), 'r')
```

This looks like a good fit, but could there be more intricacies at play? Let's try grouping by certain features.

```python
object_cols = pd.concat([housing_clean[['SalePrice', 'YrsAgoBlt']],
                         housing_clean.loc[:, (housing_clean.dtypes == object)]], axis=1)
```

```python
object_cols
```

```python
for col in object_cols.columns:
    if col == 'SalePrice':
        continue
    print(object_cols[col].value_counts())
```

```python
sb.boxplot(data=housing_clean, x='MSZoning', y='YrsAgoBlt')
```

```python
sb.boxplot(data=housing_clean, x='HouseStyle', y='YrsAgoBlt')
```

```python
sb.scatterplot(data=housing_clean, x='YrsAgoBlt', y='SalePrice', hue='HouseStyle')
```

```python
pd.get_dummies(object_cols['HouseStyle']).drop(columns=['1Story'])
```

```python
X = pd.concat([object_cols['YrsAgoBlt'],
               pd.get_dummies(object_cols['HouseStyle']).drop(columns=['1Story'])], axis=1)
X
```

```python
y = np.log(object_cols['SalePrice'])
y
```

```python
lm2 = LinearRegression()
lm2.fit(X, y)
[lm2.coef_, lm2.intercept_]
```

```python
[lm.coef_, lm.intercept_]
```

```python
lm2_coefs = pd.DataFrame(data=lm2.coef_, index=lm2.feature_names_in_, columns=['coefs'])
lm2_coefs
```

$$\ln P = -0.00847Y + 0.138X_{1.5Fin} - 0.0688X_{1.5Unf} + 0.841X_{2.5Fin} + 0.498X_{2.5Unf} + 0.118X_{2Story} - 0.173X_{SFoyer} - 0.0107X_{SLvl} + 12.322$$  
$$P = 224582\exp{(-0.00847Y + 0.138X_{1.5Fin} - 0.0688X_{1.5Unf} + 0.841X_{2.5Fin} + 0.498X_{2.5Unf} + 0.118X_{2Story} - 0.173X_{SFoyer} - 0.0107X_{SLvl})}$$

```python
g = sb.FacetGrid(housing_clean, col='HouseStyle', height=3.5)
g.map(sb.scatterplot, 'YrsAgoBlt', 'SalePrice').set(yscale='log')
```

```python
housing_clean['HouseStyle'].value_counts()
```

```python
lm_1story = LinearRegression()
lm_1story.fit(housing_clean[housing_clean['HouseStyle'] == '1Story'][['YrsAgoBlt']],
              np.log(housing_clean[housing_clean['HouseStyle'] == '1Story']['SalePrice']))

lm_2story = LinearRegression()
lm_2story.fit(housing_clean[housing_clean['HouseStyle'] == '2Story'][['YrsAgoBlt']],
              np.log(housing_clean[housing_clean['HouseStyle'] == '2Story']['SalePrice']))
```

```python
lm_1story.score(housing_clean[housing_clean['HouseStyle'] == '1Story'][['YrsAgoBlt']],
              np.log(housing_clean[housing_clean['HouseStyle'] == '1Story']['SalePrice']))
```

```python
lm_2story.score(housing_clean[housing_clean['HouseStyle'] == '2Story'][['YrsAgoBlt']],
              np.log(housing_clean[housing_clean['HouseStyle'] == '2Story']['SalePrice']))
```

```python
print([lm_1story.coef_, lm_1story.intercept_])
print([lm_2story.coef_, lm_2story.intercept_])
```

```python
plt.yscale('log')

sb.scatterplot(data=housing_clean[housing_clean['HouseStyle'] == '1Story'],
               x='YrsAgoBlt', y='SalePrice', alpha=0.5)
plt.plot(range(0, 130),
         np.exp(lm_1story.intercept_)*np.power(np.exp(lm_1story.coef_[0]),range(0, 130)), 'r', linewidth=3)
```

```python
logprice_predict_1story = lm_1story.predict(housing_clean[housing_clean['HouseStyle'] == '1Story'][['YrsAgoBlt']])
logprice_predict_1story
```

```python
plt.scatter(housing_clean[housing_clean['HouseStyle'] == '1Story']['YrsAgoBlt'],
            np.log(housing_clean[housing_clean['HouseStyle'] == '1Story']['SalePrice']) - logprice_predict_1story)
```

```python
plt.hist(np.log(housing_clean[housing_clean['HouseStyle'] == '1Story']['SalePrice']) - logprice_predict_1story,
         bins=100);
```

```python
plt.yscale('log')

sb.scatterplot(data=housing_clean[housing_clean['HouseStyle'] == '2Story'],
               x='YrsAgoBlt', y='SalePrice', alpha=0.5)
plt.plot(range(0, 130),
         np.exp(lm_2story.intercept_)*np.power(np.exp(lm_2story.coef_[0]),range(0, 130)), 'r', linewidth=3)
```

```python
lm_2story_2 = LinearRegression()
lm_2story_2.fit(np.log(housing_clean[housing_clean['HouseStyle'] == '2Story'][['YrsAgoBlt']]),
                np.log(housing_clean[housing_clean['HouseStyle'] == '2Story']['SalePrice']))
```

```python
plt.xscale('log')
plt.yscale('log')

sb.scatterplot(data=housing_clean[housing_clean['HouseStyle'] == '2Story'],
               x='YrsAgoBlt', y='SalePrice', alpha=0.5)
plt.plot(range(0, 130),
         np.exp(lm_2story_2.intercept_)*np.power(range(0, 130), lm_2story_2.coef_[0]), 'r', linewidth=3)
```

```python
logprice_predict_2story = lm_2story.predict(housing_clean[housing_clean['HouseStyle'] == '2Story'][['YrsAgoBlt']])
plt.scatter(housing_clean[housing_clean['HouseStyle'] == '2Story']['YrsAgoBlt'],
            np.log(housing_clean[housing_clean['HouseStyle'] == '2Story']['SalePrice']) - logprice_predict_2story)
```

```python
plt.hist(np.log(housing_clean[housing_clean['HouseStyle'] == '2Story']['SalePrice']) - logprice_predict_2story,
        bins=100);
```

```python
logprice_predict_2story_2 = lm_2story_2.predict(np.log(housing_clean[housing_clean['HouseStyle'] == '2Story'][['YrsAgoBlt']]))
plt.scatter(np.log(housing_clean[housing_clean['HouseStyle'] == '2Story'][['YrsAgoBlt']]),
            np.log(housing_clean[housing_clean['HouseStyle'] == '2Story']['SalePrice']) - logprice_predict_2story_2)
```

```python
plt.hist(np.log(housing_clean[housing_clean['HouseStyle'] == '2Story']['SalePrice']) - logprice_predict_2story_2,
        bins=100);
```
